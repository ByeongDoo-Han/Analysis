```python
li=[]
licol=[]
for i in df.columns:
    if ':' in i:
        s = i.split(':')
        if s[-1] not in licol:
            licol.append(s[-1])
        else:
            pass
    else:
        if i not in li:
            li.append(i)
        else:
            pass
```

```python
def cleanlist(s):
    dff = pd.DataFrame(df[s].value_counts().sort_values(ascending=False)).reset_index()
    dff.columns = ['answer', 'count']
    dff = dff.iloc[:33]
    dff.loc[dff['answer']=="Self-employed <em>(a person earning income directly from one's own business, trade, or profession)</em>", 'answer'] = 'Self-employed'
    dff.loc[dff['answer']=="Freelancer <em>(a person pursuing a profession without a long-term commitment to any one employer)</em>", 'answer'] = 'Self-employed'
    d = sns.barplot(data=dff, x='count', y='answer')
    plt.style.use('ggplot')
    plt.figure(figsize=(7,5))
    plt.title(s)
    return d
```

```python
def cleanlistnom(s):
    dff = pd.DataFrame(df[s].value_counts(normalize=True).sort_values(ascending=False)).reset_index()
    dff.columns = ['answer', 'ratio']
    dff = dff.iloc[:33]
    dff.loc[dff['answer']=="Self-employed <em>(a person earning income directly from one's own business, trade, or profession)</em>", 'answer'] = 'Self-employed'
    dff.loc[dff['answer']=="Freelancer <em>(a person pursuing a profession without a long-term commitment to any one employer)</em>", 'answer'] = 'Self-employed'
    plt.figure(figsize=(7,5))
    plt.style.use('ggplot')
    plt.title(s)
    d = sns.barplot(data=dff, x='ratio', y='answer')
    return d
```

```python
def cleanlistcol(s):
    dfff = pd.DataFrame(df[df.columns[df.columns.str.contains(s+'\?')]].notnull().sum()).reset_index()
    dfff.columns=['answer', 'ratio']
    dfff.sort_values(by='ratio', ascending=False, inplace=True)
    dfff['ratio'] = dfff['ratio']/dfff['ratio'].sum()
    dfff['answer'] = dfff['answer'].str.replace(':'+s+'\?', "")
    plt.figure(figsize=(7,5))
    plt.title(s)
    d = sns.barplot(data=dfff, x='ratio', y='answer')
    return d
```

```python
def cleanlistdrop(s):
    dfff = pd.DataFrame(df[df.columns[df.columns.str.contains(s+'\?')]].notnull().sum().drop('None:'+s)).reset_index()
    dfff.columns=['answer', 'ratio']
    dfff.sort_values(by='ratio', ascending=False, inplace=True)
    dfff['ratio'] = dfff['ratio']/dfff['ratio'].sum()
    dfff['answer'] = dfff['answer'].str.replace(':'+s+'\?', "")
    plt.figure(figsize=(7,5))
    plt.title(s)
    d = sns.barplot(data=dfff, x='ratio', y='answer')
    return d
```

# <span style="color:black"><strong>How many Developers in US more than Korea?</strong></span>
-----

```python
country = pd.DataFrame(df['What country do you live in?'].value_counts()).reset_index()
country.columns=['country', 'count']
country
```

```python
usa = country[country['country'].str.contains('United States')]
kor = country[country['country'].str.contains('South Korea')]
usakor = pd.merge(usa, kor, how='outer')
plt.title('How many Developers in US more than Korea?')
sns.barplot(data=usakor, x='country', y='count')
```

```python
print("Developers in US are more than Korea's by [{0}] times.".format(usa['count'].values[0] / kor['count'].values[0]))
```

```python
developkor = pd.DataFrame(korea['What do you use Python for the most?'].value_counts(normalize=True)).reset_index()
developkor.columns = ['develop', 'ratio']
plt.title('What do you use Python for the most? - Korea')
sns.barplot(data=developkor, x='ratio', y='develop')
```

```python
developus = pd.DataFrame(US['What do you use Python for the most?'].value_counts(normalize=True)).reset_index()
developus.columns = ['develop', 'ratio']
plt.title('What do you use Python for the most? - US')
sns.barplot(data=developus, x='ratio', y='develop')
```

```python
mer = pd.merge(developus, developkor,  how='outer')
mer['kind'] = np.nan
mer['kind'].iloc[:16] = 'us'
mer['kind'].iloc[16:] = 'kor'
plt.title('What do you use Python for? - US, KOREA')
sns.barplot(data=mer, x='ratio', y='develop', hue='kind', palette = 'prism')
```

- This graph shows comparative amount of Developers in each country. (Not total amount)
- Some facts we can know
    1. 'Web development' field has the most developers in US and KOREA.
    1. 'Data analysis' field is second.
